### **Creating a database and a table**

```sql
create database if not exists tests;
```

```sql
use tests;
```

```sql
create or replace table student_tests (
 name CHAR(10), test CHAR(10), 
 score TINYINT, test_date DATE
);
```

```sql
show tables;
```

### **Querying a table**

```sql
select * from student_tests;
```

```sql
select test, max(score) as max_grade from student_tests group by test;
```

### **Getting a bigger table in a scrollable DataFrame-like format**

```sql
select * from information_schema.tables where engine <> 'PERFORMANCE_SCHEMA' limit 20;
```

### **Loading the 2016 GBP/USD dataset**

```sql
create or replace table gbpusd (
    id INT, 
    time DATETIME,
    open DOUBLE,
    high DOUBLE, 
    low DOUBLE, 
    close DOUBLE
) default character set=utf8;
```
